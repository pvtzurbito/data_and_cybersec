## Data Transformation

### Converting <code>Timestamp</code> column into a datetime format, and separating them into date and time columns.

This allows us to group the data later on in the later sections.

```python
df['date and time'] = pd.to_datetime(df['timestamp'])
```

## Basic Data Analysis

Part 1: Finding how many failed login attempts are done for each user.

```python
attempts_df = df.groupby(['username', 'status']).size().reset_index(name = 'Attempt Count')
```

```python
attempts_df[attempts_df['status'] == 'failed'].sort_values(by = 'Attempt Count', ascending = False)
```

Part 2: Isolating the top five entries on the dataset and grouping them by IP addresses trying to login.

```python
first_five_df = df[df['username'].isin(['user5', 'user1', 'user3', 'user4', 'user2'])]
```

```python
first_five_df.groupby(['ip_address', 'username']).size().reset_index(name = 'Login Attempts')
```

Part 3: Investigating <code>ip_address == '10.0.0.99'</code> for any successful login. 

```python
df[(df['ip_address'] == '10.0.0.99') & (df['status'] == 'success')]
```
